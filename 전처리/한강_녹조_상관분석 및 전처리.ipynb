```python
total_han['채수위치'].unique()
```

```python
new_col = total_han['지점명'] + "(" + total_han['채수위치'] + ")"
new_col
```

```python
total_han['채수위치'] = new_col
total_han.drop(columns = '지점명', inplace = True)
total_han
```

```python
haepyeong1 = total_han[total_han['채수위치']=='낙동강(해평)']
df_haepyeong1 = haepyeong1[haepyeong1['유해남조류 세포수 (cells/㎖)']!=0]
df_haepyeong1
```

```python
import seaborn as sns
haepyeong_factor = df_haepyeong1.iloc[:,4:]
haepyeong_corr = haepyeong_factor.corr()
fig, ax = plt.subplots(figsize=(12, 10)) # 원하는 크기로 figsize 조정
sns.heatmap(haepyeong_corr, annot=True, cmap="Blues", ax=ax)
plt.show()
```

```python
def loc_corr_day(loc):
    x = total_han[total_han['채수위치']==loc]
    x_factor = x.iloc[:,4:]
    x_corr = x_factor.corr()
    fig, ax = plt.subplots(figsize=(12, 10)) # 원하는 크기로 figsize 조정
    sns.heatmap(x_corr, annot=True, cmap="Blues", ax=ax)
    plt.title("Correlation in "+ loc, fontsize=20)
    plt.show()
```

```python
ls_location = total_han['채수위치'].unique()
ls_location
```

```python
for i in ls_location:
    loc_corr_day(i)
```

```python
# df_nakdong_day['조사일']= df_nakdong_day['조사일'].astype('str')
total_han['조사일'] = pd.to_datetime(total_han['조사일'])
total_han
```

```python
def graph(loc):
    total_han[total_han['채수위치']==loc].plot(x='조사일',y='유해남조류 세포수 (cells/㎖)', figsize = (10,5))
    plt.title('Graph of ' + loc)
```

```python
for i in ls_location:
    graph(i)
```

```python
classify = total_han['유해남조류 세포수 (cells/㎖)'].apply(lambda x:1 if x>=10000  else 0)
classify
```

```python
# 경계발령 수치 10000을 기준으로 1 과 0으로 분류
total_han['경계발령'] = classify
total_han
```
